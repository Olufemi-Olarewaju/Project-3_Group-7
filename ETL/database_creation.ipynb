```python
#Establish connection to outbreaks_db on postgres
conn = psycopg2.connect(host="localhost",
                        dbname = "outbreaks_db",
                        user="postgres",
                        password=f"{your_password}",
                        port="5432")
```

```python
#Creating a cursor object using the cursor() method
cursor = conn.cursor()
```

```python
# Create outbreaks table in outbreaks_db
cursor.execute("DROP TABLE IF EXISTS outbreaks")
query = '''CREATE TABLE "outbreaks" (
    "ID" INTEGER   NOT NULL,
    "Institution_Name" VARCHAR   NOT NULL,
    "Institution_Address" VARCHAR   NOT NULL,
    "Outbreak_Setting" VARCHAR   NOT NULL,
    "Type_of_Outbreak" VARCHAR   NOT NULL,
    "Causative_Agent" VARCHAR   NOT NULL,
    "Date_Outbreak_Began" DATE   NOT NULL,
    "Date_Declared_Over" DATE,
    "Outbreak_Status" VARCHAR   NOT NULL,
    "Longitude" DEC   NOT NULL,
    "Latitude" DEC   NOT NULL,
    CONSTRAINT "pk_Outbreak_Table" PRIMARY KEY (
        "ID"
    )
);'''
cursor.execute(query)
print("outbreaks table created successfully")
conn.commit()
```

```python
# Import cleaned_outbreaks_with_coords.csv into outbreaks table
with open(r'Resources/cleaned_outbreaks_with_coords.csv', 'r') as f:
    # Skip the header row
    next(f)  
    cursor.copy_expert("""COPY outbreaks FROM STDIN WITH (FORMAT CSV)""", f)
    print("cleaned_outbreaks_with_coords.csv imported into outbreaks table successfully")
conn.commit()
```

```python
#Create outbreaks by year & season table in outbreaks_db
cursor.execute("DROP TABLE IF EXISTS outbreaks_by_year_season")
query = '''CREATE TABLE "outbreaks_by_year_season" (
    "Outbreak Year" INT NOT   NULL,
    "Outbreak Season" VARCHAR   NOT NULL,
    "Count of Outbreaks" INT   NOT NULL
);'''
cursor.execute(query)
print("outbreaks by year/season table created successfully")
conn.commit()
```
