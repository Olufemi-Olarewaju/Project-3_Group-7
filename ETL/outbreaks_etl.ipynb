```python
import pandas as pd
```

```python
#Load csv files into dataframes
df_2016 = pd.read_csv('Resources/ob_report_2016.csv')
df_2017 = pd.read_csv('Resources/ob_report_2017.csv')
df_2018 = pd.read_csv('Resources/ob_report_2018.csv')
df_2019 = pd.read_csv('Resources/ob_report_2019.csv')
df_2020 = pd.read_csv('Resources/ob_report_2020.csv')
df_2021 = pd.read_csv('Resources/ob_report_2021.csv')
df_2022 = pd.read_csv('Resources/ob_report_2022.csv')
df_2023 = pd.read_csv('Resources/ob_report_2023.csv')
```

```python
#Check total number of rows
print(len(df_2016) + len(df_2017) + len(df_2018) + len(df_2019) + len(df_2020) + len(df_2021) + len(df_2022) + len(df_2023))
```

```python
#Rename 'Causative Agent-1' and 'Causative Agent-2' columns from df_2020 to match all other years
df_2020 = df_2020.rename(columns = {'Causative Agent - 1': 'Causative Agent-1',
                                                                                    'Causative Agent - 2': 'Causative Agent-2'})
df_2020.columns
```

```python
#Use pd.concat to combine all 7 DFs together
df_all = pd.concat([df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023], axis = 0, ignore_index = True)
df_all
```

```python
#Check DF for missing fields or null values
df_all.info()
#Some missing values in 'Date Declared Over' for currently active outbreaks
#Causative Agent-2 has over 4000 null rows
```

```python
#Drop Causative Agent-2 and rename Causative Agent-1 to Causative Agent
df_all = df_all.drop('Causative Agent-2', axis = 1)
df_all = df_all.rename(columns = {'Causative Agent-1': 'Causative Agent'})
df_all.head()
```

```python
#Change Active column to Status, Y to Active and N to Inactive
renamed_active_values = {'Y': 'Active', 'N': 'Inactive'}
df_all = df_all.rename(columns = {'Active': 'Outbreak Status'})
df_all['Outbreak Status'] = df_all['Outbreak Status'].replace(renamed_active_values)
df_all.head()
```

```python
#Drop the '_id' column as there are some duplicates from the original DFs
df_all = df_all.drop('_id', axis = 1)
df_all.head()
```

```python
#Check if any dtypes need to be converted
df_all.dtypes
```

```python
#Change Date Outbreak Began and Date Outbreak Over to datetime objects
df_all['Date Outbreak Began'] = pd.to_datetime(df_all['Date Outbreak Began'], format = '%Y-%m-%d')
df_all['Date Declared Over'] = pd.to_datetime(df_all['Date Declared Over'], format = '%Y-%m-%d')
df_all.head()
```

```python
#Check dtypes again
df_all.dtypes
```

```python
#Check Type of Outbreak values
print(df_all['Type of Outbreak'].unique())
print(df_all['Type of Outbreak'].value_counts())
```

```python
#Check Causative Agent values
df_all['Causative Agent'].unique()
```

```python
#Recategorive Causative Agent for readability and simplicity
causative_agents = {'Influenza A (H3N2)': 'Influenza A', 'Influenza A (H3)': 'Influenza A', 'Influenza A (Not subtyped)': 'Influenza A',
                    'Parainfluenza type 1': 'Parainfluenza', 'Parainfluenza type 3': 'Parainfluenza', 'Shigella flexneri (Group B)': 'Shigella flexneri',
                    'Influenza A ((H1N1)pdm09)': 'Influenza A', 'Parainfluenza type 2': 'Parainfluenza', 'Parainfluenza type 4': 'Parainfluenza',
                    'Parainfluenza Type 1': 'Parainfluenza', 'Clostridium difficile (Toxin B)': 'Clostridium difficile', 'Influenza A (H1N1)': 'Influenza A',
                    'CPE - Colonization': 'CPE', 'Clostridium difficile (Toxin A & B)': 'Clostridium difficile', 'Influenza A (H3), Parainfluenza type 2': 'Influenza A and Parainfluenza',
                    'Influenza A (H3), Parainfluenza type 1': 'Influenza A and Parainfluenza', 'Influenza A (H1)': 'Influenza', 'Parainfluenza virus': 'Parainfluenza',
                    '(CPE) - Infection - (KPC)': 'CPE', '(CPE) - Infection - (VIM)': 'CPE', '(CPE) - Unspecified (NDM)': 'CPE', '(CPE) - Colonization (KPC)': 'CPE',
                    'COVID-19': 'Coronavirus', 'Coronavirus*': 'Coronavirus', 'Parainfluenza UNSPECIFIED': 'Parainfluenza', 'CPE Unspecified (NDM)': 'CPE',
                    'CPE Unspecified (VIM)': 'CPE', 'Parainfluenza PIV III': 'Parainfluenza', 'CPE Enterobacter unspecified (NDM)': 'CPE', 'CPE Unspecified (KPC)': 'CPE',
                    'Pending': 'Unable to identify'}
df_all['Causative Agent'] = df_all['Causative Agent'].replace(causative_agents)
df_all.head()
```

```python
#Check Causative Agent values again
df_all['Causative Agent'].unique()
```

```python
#Export df_all to a CSV file
df_all.to_csv('Resources/outbreaks_cleaned_2016_2023.csv')
```

## GeoApify Address Autocomplete API

```python
#Get a list of unique institution addresses from df_all
inst_addresses = list(df_all['Institution Address'].unique())
len(inst_addresses)
```

```python
#Import dependencies
import requests
import json
from pprint import pprint

#Import API key from config file
from config import geoapify_key
```

```python
#Use Geocoding API to get Toronto limits coordinates
geo_url = f"https://api.geoapify.com/v1/geocode/search?city=Toronto&format=json&apiKey={geoapify_key}"
geo_response = requests.get(geo_url).json()
#pprint(geo_response)
toronto_filter = geo_response['results'][0]['bbox']
print(toronto_filter)
```

```python
#Test API with one address
address = inst_addresses[204]
#Parameters include a filter to limit query within the city of Toronto
#Using coordinates from toronto_filter
query_url = f"https://api.geoapify.com/v1/geocode/autocomplete?text={address}&filter=rect:{toronto_filter['lon1']},{toronto_filter['lat1']},{toronto_filter['lon2']},{toronto_filter['lat2']}&format=json&apiKey={geoapify_key}"
test_response = requests.get(query_url).json()
pprint(test_response)
#pprint(test_response['results'][0]['lat'])
```

```python
#Loop through inst_addresses, get the latitude and longitude from the API and add the address and coordinates to a dictionary
address_coordinates = {}
for i in range(0, len(inst_addresses)):
    address = inst_addresses[i]
    query_url = f"https://api.geoapify.com/v1/geocode/autocomplete?text={address}&format=json&apiKey={geoapify_key}"
    response = requests.get(query_url).json()
    latitude = response['results'][0]['lat']
    longitude = response['results'][0]['lon']
    coordinates = [latitude, longitude]
    address_coordinates[address] = coordinates
```
