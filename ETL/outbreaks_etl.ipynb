```python
#Store full institution addresses in a list
inst_addresses = list(df_all['Institution Address'].unique())
inst_addresses
```

```python
#Import regex
import re

#First pattern I tried: capture group of digit character followed by space followed by non-digit character, then followed by whitespace followed by a dash (-) 
#(trying to match the pattern '123 University Ave - Unit 3' where '123 University Ave' is the capture group)
pattern1 = re.findall(r'(\d*\s\D*)\s-\s', str(inst_addresses))
pattern1
#This works for only about 150 of the addresses as only those ones match the above pattern
```

```python
#Tried a second pattern, same as above but adding another whitespace and then a non-digit character at the end of the pattern outside of the capture group
pattern2 = re.findall(r'(\d+\s\D+)\s-\s\D', str(inst_addresses))
pattern2
#This also works for some of the addresses - about 80
```

```python
#Tried looping through the addresses and using an if statement so that if the address contains a dash, split the string at the dash and append the address to a new list
extr_addresses = []
for address in inst_addresses:
    if ' - ' in str(address):
        new_address = address.split(r'\s-\s')
        extr_addresses.append(new_address)
    else:
        print('nothing')
#This doesn't work and keeps printing "nothing" - tried different variations and none worked
```

```python
#Import dependencies
import requests
import json
from pprint import pprint

#Import API key from config file
from config import geoapify_key
```

```python
#Use Geocoding API to get the place ID for Toronto, to be used in the Address Autocomplete API parameters
geo_url = f"https://api.geoapify.com/v1/geocode/search?city=Toronto&type=city&format=json&apiKey={geoapify_key}"
geo_response = requests.get(geo_url).json()
pprint(geo_response)
toronto_place_id = geo_response['results'][0]['place_id']
#This was required when I was trying to pass the list of addresses into the API as it would sometimes return an address not within Toronto
#Using the place ID in the Address API will limit the results to the city of Toronto
```

```python
#Test API with one address
address = inst_addresses[226]
#Parameters include a filter to limit query within the city of Toronto
query_url = f"https://api.geoapify.com/v1/geocode/autocomplete?text={address}&filter=place:{toronto_place_id}&format=json&apiKey={geoapify_key}"
test_response = requests.get(query_url).json()
pprint(test_response)
#This works with some addresses but not all - for example try changing the index in to 226 and it returns nothing in the results
```

```python
#After I tried the above API request with the very first address and it worked, I thought it would work for the rest of the addresses
#So created a for loop to fetch from the API for each address in inst_addresses, appending the address as a key and the longitude/latitude as a value in a dictionary
#This runs for about 50 seconds before returning IndexError: list index out of range
#This is what made me test the above cell with different addresses to find the error, which then made me realize the issue with the address formatting

#Loop through inst_addresses, get the latitude and longitude from the API and add the address and coordinates to a dictionary
address_coordinates = {}
for i in range(0, len(inst_addresses)):
    address = inst_addresses[i]
    query_url = f"https://api.geoapify.com/v1/geocode/autocomplete?text={address}&format=json&apiKey={geoapify_key}"
    response = requests.get(query_url).json()
    latitude = response['results'][0]['lat']
    longitude = response['results'][0]['lon']
    coordinates = [latitude, longitude]
    address_coordinates[address] = coordinates
```
