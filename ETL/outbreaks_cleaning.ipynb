```python
#Import dependencies for API
import requests
import json
from pprint import pprint

#Import API key from config file
from config import geoapify_key
```

```python
#Use Geocoding API to get the place ID for Toronto, to be used in the Address Autocomplete API parameters
#Using the place ID in the Address API will limit the results to the city of Toronto
geo_url = f"https://api.geoapify.com/v1/geocode/search?city=Toronto&type=city&format=json&apiKey={geoapify_key}"
geo_response = requests.get(geo_url).json()
pprint(geo_response)
toronto_place_id = geo_response['results'][0]['place_id']
```

```python
#Separate cleaned_addresses into 4 smaller lists
clean_add_1 = cleaned_addresses[0:133]
clean_add_2 = cleaned_addresses[133:266]
clean_add_3 = cleaned_addresses[266:399]
clean_add_4 = cleaned_addresses[399:534]
```

```python
#API call for first list of addresses
clean_dict = {}
failed_addresses = []

for i in range(0, len(clean_add_1)):
    address = clean_add_1[i]
    address_url = f"https://api.geoapify.com/v1/geocode/autocomplete?text={address}&filter=place:{toronto_place_id}&format=json&apiKey={geoapify_key}"
    address_response = requests.get(address_url).json()
    try:
        latitude = address_response['results'][0]['lat']
        longitude = address_response['results'][0]['lon']
        coords = [longitude, latitude]
        clean_dict[address] = coords
        print(f'API call successful for index {i}')
    except IndexError:
        failed_addresses.append(address)
        print(f'API call failed for index {i}')
```

```python
#API call for second list of addresses

for i in range(0, len(clean_add_2)):
    address = clean_add_2[i]
    address_url = f"https://api.geoapify.com/v1/geocode/autocomplete?text={address}&filter=place:{toronto_place_id}&format=json&apiKey={geoapify_key}"
    address_response = requests.get(address_url).json()
    try:
        latitude = address_response['results'][0]['lat']
        longitude = address_response['results'][0]['lon']
        coords = [longitude, latitude]
        clean_dict[address] = coords
        print(f'API call successful for index {i}')
    except IndexError:
        failed_addresses.append(address)
        print(f'API call failed for index {i}')
```

```python
#Graphing time series by season
df = pd.read_csv('Resources/cleaned_outbreaks_with_coords.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
#Change Date Outbreak Began and Date Outbreak Over to datetime objects
df['Date Outbreak Began'] = pd.to_datetime(df['Date Outbreak Began'], format = '%Y-%m-%d')
df['Date Declared Over'] = pd.to_datetime(df['Date Declared Over'], format = '%Y-%m-%d')
df.head()
```

```python
#Add columns for year and month
df['Outbreak Year'] = df['Date Outbreak Began'].dt.strftime('%Y')
df['Outbreak Month'] = df['Date Outbreak Began'].dt.strftime('%m')
df.head()
```

```python
#Convert int month to name month
df['Outbreak Month'] = pd.to_datetime(df['Outbreak Month'], format='%m').dt.month_name()
df.head()
```

```python
#Create dictionary to map seasons
seasons_dict = {'January': 'Winter', 
                'February': 'Winter', 
                'March': 'Spring', 
                'April': 'Spring', 
                'May': 'Spring', 
                'June': 'Summer',
                'July': 'Summer',
                'August': 'Summer',
                'September': 'Fall',
                'October': 'Fall',
                'November': 'Fall',
                'December': 'Winter'}

#Use .apply with lambda x to add a column to df for the season
df['Outbreak Season'] = df['Outbreak Month'].apply(lambda x: seasons_dict[x])
df.head()
```

```python
df.to_csv('Resources/outbreaks.csv')
```

```python
#Group DF by outbreak season and get count of outbreaks for all years
all_seasons = df.groupby('Outbreak Season', as_index = False)['Institution Name'].count()
all_seasons['Outbreak Year'] = 'All Years'
all_seasons = all_seasons[['Outbreak Year', 'Outbreak Season', 'Institution Name']]
all_seasons
```

```python
#Group DF by year and season and get the count of outbreaks
grouped_seasons = df.groupby(['Outbreak Year', 'Outbreak Season'], as_index = False)['Institution Name'].count()
grouped_seasons
```

```python
grouped_seasons_all = pd.concat([all_seasons, grouped_seasons], axis = 0, ignore_index = True)
grouped_seasons_all
```
